```python
#crear lista y bandera como falso
lista_de_alumnos  = []
lista_edades_alumnos = []
lista_promedio_alumnos = []


bandera_rut = False
bandera_registro = False
contador = 0
indice = 0
```

```python
#si no es falso bandera_registro
#importante
while not bandera_registro:
  try:
    
    rut = input("Ingrese rut de alumno con puntos y guion:\n")
    edad = int(input("Ingrese la edad del alumno {} ".format (rut)))
    bandera_registro = True

  except:
    print("ocurrio un error")
    bandera_registro = False
```

```python
#validaciones

if rut != " ":
  #validar rut
  for letra in rut:
    if letra  == "." or letra  == "-":
        bandera_rut = True
```

```python
#/t tabulacion
#implementando listas
if bandera_rut:
  print(f"Datos alumnos\n \t {rut},{edad}")
  lista_de_alumnos.append(rut)
  lista_edades_alumnos.append(edad)
  lista_promedio_alumnos.append(0.0)
```

```python
rutBuscado = "24.555.666-7"
for r in lista_de_alumnos:
  if r == rutBuscado:
    print("Rut existe")
    indice = contador
    contador = contador + 1
```

```python
print("RUT: {}".format(lista_de_alumnos[indice]))
print("EDAD: {}".format(lista_edades_alumnos[indice]))
print("PROMEDIO: {}".format(lista_promedio_alumnos[indice]))
```
