```python
#Mad Libs (Mejorada)

#Declaramos las variables

i = 0

#Damos la bienvenida mostrando con print

print("========================================")
print("    Bienvenido(a) a Historias Locas")
print("========================================")

#Con un While podemos hacer que se repitan las 3 historias

while i < 5:
  tema = input("\nIngresa que tema quieres para tu historia:\n1) Comida.\n2) Ropa.\n3) Musica.\nTema: ")
  
#Empezamos con el primer tema sobre comida

  if tema == "1":
    color1 = input("\nIngresa las palabras que faltan para completar la historia\nColor 1 (En plural): ")
    color2 = input("Color 2 (En plural): ")
    adjetivo = input("Adjetivo (plural): ")
    fruta = input("Fruta (plural): ")
    print(f"\n¡La historia ahora si cobra sentido!:\n\nCuando crezcamos seremos {color1} y {color2}, crecemos en los arboles, somos nutritivas, {adjetivo} y también puedes comernos\nsin que estemos cocinadas. Por eso somos las {fruta}")
    i = i + 1

    if i < 3:
      ronda = input("\n\n\n¿Quieres jugar otra ronda?\n1) Si\n2) No\nOtra ronda: ")

      if ronda == "1":
        i = i + 1
      
      if ronda == "2":
        i = i + 5

#Empezamos con el segundo tema sobre ropa

  if tema == "2":
    numero = input("\nIngresa las palabras que faltan para completar la historia\nNumero: ")
    estacion = input("Estacion del año: ")
    estacion2 = input("Estacion del año 2: ")
    print(f"\n¡La historia ahora si cobra sentido!:\n\nTienen justo {numero} dedos como la mano; se rellenan en {estacion}, se vacían en {estacion2}. Mucho gusto, somos los guantes.")
    i = i + 1

    if i < 4:
      ronda = input("\n\n\n¿Quieres jugar otra ronda?\n1) Si\n2) No\nOtra ronda: ")

      if ronda == "1":
        i = i + 1

      if ronda == "2":
        i = i + 5

#Empezamos con el tercer tema sobre ropa

  if tema == "3":
    adjetivo = input("\nIngresa las palabras que faltan para completar la historia\nAdjetivo (En singular): ")
    parte = input("Parte del cuerpo: ")
    adjetivo2 = input("Adjetivo en femenino (Singular): ")
    print(f"\n¡La historia ahora si cobra sentido!:\n\nSoy {adjetivo} y de madera, tengo un arco y no flecha. Me pones entre hombro y {parte} y mi música es {adjetivo2}")
    i = i + 1

    if i < 4:
      ronda = input("\n\n\n¿Quieres jugar otra ronda?\n1) Si\n2) No\nOtra ronda: ")

      if ronda == "1":
        i = i + 1

      if ronda == "2":
        i = i + 5

print("\nNo mas rondas, hasta luego\n\n\n")
```
