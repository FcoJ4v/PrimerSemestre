```python
import random
# Mad Libs (Historias Locas)
from random import choice
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

def historia_aleatoria (historias):
  try:
      with open(historias, 'r', encoding='utf-8') as f:
        lineas = f.read().splitlines()
        return choice(lineas)
  except FileNotFoundError:
      return None

nombre_archivo = '/content/drive/My Drive/historias.txt'
resultado  = historia_aleatoria(nombre_archivo)
resultado1 = historia_aleatoria(nombre_archivo)

print('Bienvenio a Mads Libs!\n')


cant_jugador = int(input('Ingresa la cantidad de jugadores:\n '))

if(cant_jugador == 1):

  jugador1 = input('Ingresa el nombre del primer jugador: \n')

 
elif(cant_jugador>1 and cant_jugador<=2):
  jugador1 = input('Ingresa el nombre del primer jugador: \n')
  jugador2 = input('Ingresa el nombre del segundo jugador: \n')

  
  print(f'{jugador1} esta es tu historia:\n {resultado}')
  
  print(f'{jugador2} esta es tu historia:\n {resultado1}')

  
elif(cant_jugador>=1 and cant_jugador<=3):


  jugador1 = input('Ingresa el nombre del primer jugador: \n')
  
  jugador2 = input('Ingresa el nombre del segundo jugador: \n')
  
  jugador3 = input('Ingresa el nombre del tercer jugador: \n')
  
  print(f'{jugador1} esta es tu historia:\n {resultado}') 
  print(f'{jugador2} esta es tu historia:\n {resultado1}')
  print(f'{jugador3} esta es tu historia:\n {resultado}')
```
