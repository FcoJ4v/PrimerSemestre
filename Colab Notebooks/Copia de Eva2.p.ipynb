1. Una emprendedora comenzó a vender identificaciones para mascotas con entrega a domicilio. Estas son metálicas con diversas formas. Para ello creó una tienda online donde los clientes solicitan los productos y la grabación de la placa metálica. Cada identificación tiene un precio, según el tamaño de la placa:

●	Pequeña: $3.500
●	Mediana: $4.500
●	Grande: $5.500


El despacho a domicilio tiene un recargo de $2000 y si la compra es superior a 10.000 el despacho es gratis.

Determine el total a pagar por un cliente que requiere una cantidad X de identificaciones para sus mascotas.



```python
#recordar ver tema de elegir diferentes tipos de placas y mostrarlas en el detalle de la venta con su precio unitario.

placa_p = 3500
placa_m = 4500
placa_g = 5500

print('Bienvenido al sistema de ventas.\n')

seleccion = int(input('Seleccione el tamaño de la placa deseada:\n 1) Pequeño: $3.500\n 2) Mediana: $4.500\n 3) Grande: $5.500\n'))
cantidad = int(input('Ingrese la cantidad deseada: \n'))
#flag_tipo = False
#tipo =  int(input('¿Desea agregar otro tipo de placa?\n 1)SI \n 2)NO\n'))
#if(tipo == 1):
 # flag_tipo = True
grabacion_placa = int(input('¿Desea agregar una grabacion en la placa?\n 1)SI \n 2)NO\n'))
despacho = int(input('¿Desea despacho  domicilio?\n 1)SI \n 2)NO\n'))

#Validacion y calculo placa pequeña
if(seleccion ==1 and grabacion_placa ==1 and despacho == 1):
  prec_cant = (cantidad * placa_p)
  if(prec_cant>10000):
    print(f'Detalle Compra:\n-Valor placa: ${placa_p}\n-Cantidad: {cantidad}\n-Despacho: $0 \n-El total de su compra es: ${prec_cant}\n')
  else:
    total = (prec_cant + 2000)
    print(f'Detalle Compra:\n-Valor placa: ${placa_p}\n-Cantidad: {cantidad}\n-Despacho: $2000 \n-El total de su compra es: ${total}\n')


#Validacion y calculo placa mediana
elif (seleccion ==2 and grabacion_placa ==1 and despacho == 1):
  prec_cant = (cantidad * placa_m)
  if(prec_cant>10000):
    print(f'Detalle Compra:\n-Valor placa: ${placa_m}\n-Cantidad: {cantidad}\n-Despacho: $0 \n-El total de su compra es: ${prec_cant}\n')
  else:
    total = (prec_cant + 2000)
    print(f'Detalle Compra:\n-Valor placa: ${placa_m}\n-Cantidad: {cantidad}\n-Despacho: $2000 \n-El total de su compra es: ${total}\n')

#Validacion y calculo placa grande
elif (seleccion ==3 and grabacion_placa ==1 and despacho == 1):
  prec_cant = (cantidad * placa_g)
  if(prec_cant>10000):
    print(f'Detalle Compra:\n-Valor placa: ${placa_g}\n-Cantidad: {cantidad}\n-Despacho: $0 \n-El total de su compra es: ${prec_cant}\n')
  else:
    total = (prec_cant + 2000)
    print(f'Detalle Compra:\n-Valor placa: ${placa_g}\n-Cantidad: {cantidad}\n-Despacho: $2000 \n-El total de su compra es: ${total}\n')
else:
  print('Opciones Ingresadas no validas.')

```
