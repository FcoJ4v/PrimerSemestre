```python
zap_valor = 20000
total_compra = 0

cant_zap = int(input('Ingrese la cantidad de zapatos que quiere comprar: '))
total_compra = (zap_valor * cant_zap)

if total_compra >= 40000:
  print(f"El total de la compra realizada es:\n${total_compra},", "el envio es gratuito")
else:
  calc_comp_envio = (total_compra + 3000)
  print (f"El total de la compra realizada es:\n${calc_comp_envio}")
```
